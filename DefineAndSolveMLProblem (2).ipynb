## Preparing Data

```python
df = df.dropna()
#drop nulls
```

```python
df.rename(columns = {'Freedom to make life choices': 'Freedom'}, inplace = True)
df
```

```python
features = ['Life Ladder', 'Social support', 'Freedom']
```

```python
#winsorizing
for feature in features:
    df[feature] = stats.mstats.winsorize(df[feature], limits=[0.01, 0.01])

df['Positive affect'] = stats.mstats.winsorize(df['Positive affect'], limits=[0.01, 0.01])
```

```python
#defining variables
y = df['Positive affect']
X = df[['Life Ladder', 'Social support', 'Freedom']]
```

```python
df['Positive affect'].shape
```

## Model Training

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Create the LinearRegression model object 
model = LinearRegression()

# Fit the model 
model.fit(X_train, y_train)

#  Make predictions on the test data 
prediction = model.predict(X_test)
```

## Examining results

```python
mse = mean_squared_error(y_test, prediction)
print("Mean Squared Error on Test Set:", mse)
```

```python
print('Model Summary:\n')

# Print intercept (alpha)
print('Intercept:')
print('alpha = ' , model.intercept_)

# Print weights
print('\nWeights:')
i = 0
for w in model.coef_:
    print('w_',i+1,'= ', w, ' [ weight of ', features[i],']')
    i += 1
```

```python
print(' R^2 =   %.2f'
      % r2_score(y_test, prediction))
```

## K Fold Cross Validation

```python
kf = KFold(n_splits=5, random_state=None)
cv_scores = cross_val_score(model, X, y, cv=kf, scoring='neg_mean_squared_error')
cv_mse = -cv_scores.mean()
print("Cross-Validation Mean Squared Error:", cv_mse)
```

## Regularization

```python
from sklearn.linear_model import Lasso
```

```python
#Training it
lasso = Lasso(alpha=1.0)


lasso.fit(X_train, y_train)

# Predict and evaluate
lasso_pred = lasso.predict(X_test)
lasso_mse = mean_squared_error(y_test, lasso_pred)
print("Lasso Regression MSE:", lasso_mse)
```

```python
param_grid = {'alpha': [0.1, 1.0, 10.0, 100.0]}
grid_search = GridSearchCV(Lasso(), param_grid, cv=kf, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
best_alpha = grid_search.best_params_['alpha']
print("Best alpha for Lasso:", best_alpha)
```
